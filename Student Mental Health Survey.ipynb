```python
def convert_cgpa(cgpa):
    if '-' in cgpa:
        low, high = map(float, cgpa.split('-'))
        return (low + high) / 2
    return float(cgpa)

df['cgpa'] = df['cgpa'].apply(convert_cgpa)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
features = df[['depression', 'anxiety', 'isolation', 'future_insecurity']]

scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)  
df['mental_health_cluster'] = kmeans.fit_predict(features_scaled)
```

```python
sns.scatterplot(data=df, x='depression', y='anxiety', hue='mental_health_cluster', palette='Set1')
plt.title('Clustering Students by Mental Health')
plt.show()
```

## Clusters
Cluster 1: Low depression, low anxiety â€” potentially less stressed students.

Cluster 2: High anxiety, moderate depression â€” moderate stress group.

Cluster 3: High across all indicators â€” high-stress group.
